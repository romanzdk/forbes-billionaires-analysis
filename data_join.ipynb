```python
import pandas as pd
import os
import glob
```

```python
mypath = "."

# get list of files
files = [f for f in os.listdir(mypath) if (os.path.join(mypath, f))[-4:] == "json"]

# build list of dataframes, adding "challenge" column
dfs = [pd.read_json(os.path.join(mypath, f)).assign(year=f[:4]) for f in files]

# concatenate dataframes into one
df = pd.concat(dfs, ignore_index=True)

# write to csv
df.to_csv('all_entrants.csv', index = False)
```
