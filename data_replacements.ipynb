```
import pandas as pd
import numpy as np
```

```
df = pd.read_csv("/content/drive/My Drive/_Stahovani/all_entrants.csv")
```

```
df['industry'] = df['industry'].replace(['Finance and Investments', 'Finance', 'Banks', 'Investments'], 'Finance & Investments')
df['industry'] = df['industry'].replace(['Fashion and Retail', 'Retail', 'Fashion'], 'Fashion & Retail')
df['industry'] = df['industry'].replace(['Food and Beverage', 'Food', 'Beverages'], 'Food & Beverage')
df['industry'] = df['industry'].replace(['Media', 'Celebrity'], 'Media & Entertainment')
df['industry'] = df['industry'].replace(['Health care', 'Health Care', 'Medicine'], 'Healthcare')
df['industry'] = df['industry'].replace(['Information Technology', 'Software', 'Electronics', 'Internet'], 'Technology')
df['industry'] = df['industry'].replace(['Telecom'], 'Telecommunications')
df['industry'] = df['industry'].replace(['Oil'], 'Energy')
df['industry'] = df['industry'].replace(['Transportation', 'Shipping'], 'Logistics')
df['industry'] = df['industry'].replace(['Gaming', 'Casinos &#38; Gaming'], 'Gambling & Casinos')
df['industry'] = df['industry'].replace(['Metals &#38; Mining', 'Coal', 'Steel'], 'Metals & Mining')
```

```
df.loc[df['name'].str.contains('&'), 'gender'] = 'Family'
df.loc[df['name'].str.contains(' and '), 'gender'] = 'Family'
df.loc[df['name'].str.contains('family'), 'gender'] = 'Family'
df.loc[df['name'].str.contains('Boris Jordan'), 'gender'] = 'M'
```

```
df.loc[df['country'] == 'Czechia', 'country'] = 'Czech Republic'
df.loc[df['country'] == 'Macao', 'country'] = 'Macau'
```

```
maxY = df.groupby(['name'], as_index=False)['year'].max()
```

```
merged = pd.merge(df, maxY, on = ['name', 'year'])
```

```
maxI = merged.groupby('name', as_index = False)['industry'].min()
```

```
maxI.columns = ['name', 'industry_max_year']
```

```
result = df.merge(maxI, on = 'name')
```

```
result.to_csv('/content/drive/My Drive/Misc/billionaires/all_entrants.csv', index = False)
```
